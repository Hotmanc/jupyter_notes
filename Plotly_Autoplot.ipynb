```python
# Hotman Christianto
# Code for Plotly_Plotter.py

import logging
import numpy as np
import pandas as pd
import plotly
import plotly.graph_objects as go
from io import StringIO
from plotly.subplots import make_subplots

""" 
This function iterates through the dataframe, plotting one column against other columns in that dataframe. The
output will be 2 columns of line graphs, where the separator is defined by rw. 

Inputs:
files: The .csv files to read from. If multiple files were given, it will iterate through all of them. Note that
        only .csv files will be processed. Other files will not result in an error, but will not be processed. It
        is given by csvFile from the widgets below.
        
rw   : The column separator. Column numbers less than rw will be plotted on the left side, while those greater
        will be plotted on the right side. Given by ySep from the widgets below. 
"""

def main(rw, files = None):
    if files == None:
        raise Exception('No file or dataset is given.')
    
    for fl in range(len(files.metadata)):
        fileName = files.metadata[fl]['name']
        fileType = '.'+fileName.split('.')[-1] 
        
        if fileType == '.csv':
            
            print('Plots obtained from ' + fileName)
            # Decode the data from the file and put into a data frame.
            decode_file = files.data[fl].decode('utf-8')
            dataset = pd.read_csv(StringIO(decode_file))

            colnm = list(dataset.columns)
            xAxis = colnm[0]
            lenCol = len(dataset.columns)
            colnm.remove(colnm[0])

            # Titles for each individual plots
            fig = make_subplots(rows=rw, cols=2)

            r = 1
            c = 1

            for cn in colnm:
                fig.append_trace(go.Scatter(
                                x = dataset[xAxis],
                                y = dataset[cn],
                                name = cn), 
                                row = r, col = c,
                                )
                

                # Add labels to x and y axes.
                fig.update_xaxes(title_text = xAxis,
                                row = r, col = c)
                fig.update_yaxes(title_text = cn,
                                row = r, col = c)

                r = r+1
                if r > rw:
                    r = 1
                    c = c+1

            fig.update_layout(height = 800)
            fig.show()
            
        else:
            print(fileName + ' is not a .csv file. No plots can be made.')
```
